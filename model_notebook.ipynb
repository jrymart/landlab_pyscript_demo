```python
from hillslope_lem import HillslopeLem
import matplotlib.pyplot as plt
```

# Introduction
In this notebook we run a simple hillslope diffusion model for 1,000,000 years with various diffusion rates.  We then plot the middle swath of topography at the end of the run

```python
diffusion_coeffs = [0.0001, 0.001, 0.01, 0.1]
```

```python
fig, ax = plt.subplots()
for D in diffusion_coeffs:    
    params = HillslopeLem.DEFAULT_PARAMS
    params["diffuser"]["D"] = D
    params["output"]["plot_to_file"] = False
    params["output"]["save_times"] = [1000001]
    model = HillslopeLem(params)
    model.run()
    middle_swath = model.grid.at_node["topographic__elevation"].reshape(model.grid.shape)[:,2]
    ax.plot(middle_swath,
           label="D=%f" % D)
ax.legend()
```
